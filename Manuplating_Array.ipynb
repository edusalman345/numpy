Import

```python
import numpy as np
```

Data

```python
marks = np.array([
    [78, 85, 90],
    [45, 50, 40],
    [88, 92, 84],
    [60, 65, 58],
    [95, 98, 100]
])
```

Calculating Average marks per student

```python
avg_per_student = np.mean(marks, axis=1)
```

```python
print('Average marks per student:')
for index, x in enumerate(avg_per_student, start=1):
    print(f"Student{index} = {round(x,2)}")
```

Calculating Average marks per subject

```python
avg_per_subject = np.mean(marks, axis=0)
```

```python
print('Average marks per subject:')
for index, x in enumerate(avg_per_subject, start=1):
    print(f"Subject{index} = {round(x,2)}")
```

```python
print("Total Per Student Marks:")
student_marks = np.sum(marks, axis=1)
for index, x in enumerate(student_marks, start=1):
    print(f"Student{index} = {x}")
```

Calculating Highest Marks

```python
max_marks = student_marks.max()
student_index = np.where(student_marks == max_marks)
print(f"Topper: Student{student_index[0][0]+1} with {max_marks}")
```

Calculating Lowest Marks

```python
min_marks = student_marks.min()
student_index = np.where(student_marks == min_marks)
print(f"Lowest Marks: Student{student_index[0][0]+1} with {min_marks}")
```

Calculating Who passed > 70

```python
passed_sudents = np.where(avg_per_student > 70)
print("Passed Students:")
for index in passed_sudents[0]:
    print(f"Student{index+1}")

print(passed_sudents)
```

Calculating Who Failed < 50

```python
failled_students = np.where(marks < 50)
print("Failled Students:")
for index in failled_students[0]:
    print(f"Student{index+1}")
```

Bonous Marks (5)

```python
print('Before')
for x in marks:
  print(x)
print('After')
bonous_marks = 5
for i in range(len(marks)):
  for j in range(len(marks[i])):
    if marks[i][j] + bonous_marks <=100:
      marks[i][j] = marks[i][j] + bonous_marks
    elif marks[i][j] + bonous_marks > 100:
      marks[i][j] = 100

for x in marks:
  print(x)
```
